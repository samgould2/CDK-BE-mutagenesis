# Preprocessing data for editing analysis

- Putting things into a centralized dataframe
- Adding a naming schema

```python
import numpy as np 
import pandas as pd
import os
```

```python
kk = os.listdir('/Volumes/sanchezrivera/samgould/241105SanA/D24-257012-2-6959F_guide_split_BARCODES_redo')
kk
```

```python
#load in the editing files
cdk7 = pd.read_csv('edit_libraries/CDK7_edits.csv').rename(columns = {'gRNA_id':'gRNA_id_OLD'})
cdk8 = pd.read_csv('edit_libraries/CDK8_edits.csv').rename(columns = {'gRNA_id':'gRNA_id_OLD'})
cdk9 = pd.read_csv('edit_libraries/CDK9_edits.csv').rename(columns = {'gRNA_id':'gRNA_id_OLD'})
cdk19 = pd.read_csv('edit_libraries/CDK19_edits.csv').rename(columns = {'gRNA_id':'gRNA_id_OLD'})
```

```python
full_lib = pd.read_csv('CDK_library_CLUSTER.csv')
full_lib = full_lib[['gRNA_id', 'Gene',  'classification', 'Protospacer',  'CBE_amenable','ABE_amenable', 'gRNA_id_OLD']]
full_lib_targeting = full_lib[full_lib['classification']=='targeting']
full_lib_targeting
```

```python
#pulling out CBE edits

cdk7_g = full_lib_targeting[full_lib_targeting['Gene']=='CDK7']
cdk8_g = full_lib_targeting[full_lib_targeting['Gene']=='CDK8']
cdk9_g = full_lib_targeting[full_lib_targeting['Gene']=='CDK9']
cdk19_g = full_lib_targeting[full_lib_targeting['Gene']=='CDK19']

cdk7_merge = pd.merge(cdk7_g, cdk7[cdk7['Editor']=='CBE'], on='gRNA_id_OLD')
cdk8_merge = pd.merge(cdk8_g, cdk8[cdk8['Editor']=='CBE'], on='gRNA_id_OLD')
cdk9_merge = pd.merge(cdk9_g, cdk9[cdk9['Editor']=='CBE'], on='gRNA_id_OLD')
cdk19_merge = pd.merge(cdk19_g, cdk19[cdk19['Editor']=='CBE'], on='gRNA_id_OLD')


#then concatenate and create a new naming schema
concated = pd.concat((cdk7_merge, cdk8_merge, cdk9_merge, cdk19_merge)).reset_index(drop=True)

unique_id = []
for i, val in concated.iterrows():
    gid = val['gRNA_id']
    unique_id.append(f'{gid}_{i+1}')

concated['unique_id'] = unique_id

#and then save it
concated.to_csv('CDK_edits_CBE_pool1.csv', index=False)
```
