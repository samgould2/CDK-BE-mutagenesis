```python
import pandas as pd 
import numpy as np
```

```python
colony_seq = pd.read_csv('colony_seq_analysis.csv')
library = pd.read_csv('/Users/samgould/Documents/GitHub/CDK9-mutagenesis/CDK_library_final.csv')
```

```python
library
```

```python
proto_id = []
bc_id = []
proto_pool = []
bc_pool = []
bc_predicted = []
for i, val in colony_seq.iterrows():
    p = val['Protospacer']
    b = val['Barcode']

    proto = library.loc[library['proto_G+19']==p, 'gRNA_id']
    bc = library.loc[library['Hamming_BC']==b, 'gRNA_id']

    if len(proto)!=0:
        proto_id.append(proto.values[0])
        proto_pool.append(library.loc[library['proto_G+19']==p, 'Pool'].values[0])
        bc_predicted.append(library.loc[library['proto_G+19']==p, 'Hamming_BC'].values[0])
    elif len(proto)==0:
        proto_id.append(None)
        proto_pool.append(None)
        bc_predicted.append(None)

    if len(bc)!=0:
        bc_id.append(bc.values[0])
        bc_pool.append(library.loc[library['Hamming_BC']==b, 'Pool'].values[0])
    elif len(bc)==0:
        bc_id.append(None)
        bc_pool.append(None)
```

```python
colony_seq['barcode_predicted'] = bc_predicted
colony_seq['proto_id'] = proto_id
colony_seq['bc_id'] = bc_id
colony_seq['proto_pool'] = proto_pool
colony_seq['bc_pool'] = bc_pool


match = []

for i, val in colony_seq.iterrows():
    if val['proto_id']==val['bc_id']:
        match.append(True)
    else:
        match.append(False)

colony_seq['proto_bc_match'] = match
colony_seq
```

# Primordium identification

```python
primordium = pd.read_csv('primordium_seqs.csv')

primordium['Protospacer'] = [i.upper() for i in primordium['Protospacer']]
primordium['Sensor'] = [i.upper() for i in primordium['Sensor']]
primordium['Barcode'] = [i.upper() for i in primordium['Barcode']]

primordium
```

```python
proto_id = []
bc_id = []
proto_pool = []
sensor_id = []
bc_pool = []
bc_predicted = []
for i, val in primordium.iterrows():
    p = val['Protospacer']
    b = val['Barcode']
    s = val['Sensor']

    proto = library.loc[library['proto_G+19']==p, 'gRNA_id']
    bc = library.loc[library['Hamming_BC']==b, 'gRNA_id']
    sensor = library.loc[library['sensor_wt']==s, 'gRNA_id']

    if len(proto)!=0:
        proto_id.append(proto.values[0])
        proto_pool.append(library.loc[library['proto_G+19']==p, 'Pool'].values[0])
        bc_predicted.append(library.loc[library['proto_G+19']==p, 'Hamming_BC'].values[0])
    elif len(proto)==0:
        proto_id.append(None)
        proto_pool.append(None)
        bc_predicted.append(None)

    if len(sensor)!=0:
        sensor_id.append(sensor.values[0])
        
    elif len(sensor)==0:
        sensor_id.append(None)
        

    if len(bc)!=0:
        bc_id.append(bc.values[0])
        bc_pool.append(library.loc[library['Hamming_BC']==b, 'Pool'].values[0])
    elif len(bc)==0:
        bc_id.append(None)
        bc_pool.append(None)
```

```python
primordium['barcode_predicted'] = bc_predicted
primordium['proto_id'] = proto_id
primordium['bc_id'] = bc_id
primordium['sensor_id'] = sensor_id
primordium['proto_pool'] = proto_pool
primordium['bc_pool'] = bc_pool


match = []

for i, val in primordium.iterrows():
    if val['proto_id']==val['bc_id']:
        match.append(True)
    else:
        match.append(False)

primordium['proto_bc_match'] = match
primordium
```
