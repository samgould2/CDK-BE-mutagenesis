```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('titer_2024_09_12.csv')[:-2]

df = df.rename(columns = {'Unnamed: 0': 'sample_name', 'singlets/singlets2/singlets3/RFP+ | Freq. of Parent (%)':'RFP+ %'})[['sample_name', 'RFP+ %']]
cell_type = []
virus_amt = []
for i in df['sample_name']:
    sp = i.split('_')
    cell_type.append(sp[2])
    sp2 = int(sp[3].split('.')[0])
    virus_amt.append(sp2)

df['cell type'] = cell_type
df['Virus Added (uL) in 3mL total'] = virus_amt


A2780 = df[df['cell type']=='A2780']
A549 = df[df['cell type']=='A549']
```

```python
df2 = pd.read_csv('titer_2_2024_09_16.csv')[:-2]

df2 = df2.rename(columns = {'Unnamed: 0': 'sample_name', 'singlets/singlets2/singlets3/RFP+ | Freq. of Parent (%)':'RFP+ %'})[['sample_name', 'RFP+ %']]
cell_type = []
virus_amt = []
for i in df2['sample_name']:
    sp = i.split('_')
    cell_type.append(sp[2])
    sp2 = int(sp[3].split('.')[0])
    virus_amt.append(sp2)

df2['cell type'] = cell_type
df2['Virus Added (uL) in 2mL total'] = virus_amt

df2

```

```python
fig, ax = plt.subplots(figsize=(6,5))
ax.scatter(A549['Virus Added (uL) in 3mL total'], A549['RFP+ %'], label='A549', color='tab:red')
ax.scatter(A2780['Virus Added (uL) in 3mL total'], A2780['RFP+ %'], label='A2780', color='tab:blue')
ax.scatter(df2['Virus Added (uL) in 2mL total'], df2['RFP+ %'], label='A2780\n(with polybrene\nin 2 mL)', color='lightblue')


ax.plot(A2780['Virus Added (uL) in 3mL total'], A2780['RFP+ %'], color='tab:blue')
ax.plot(A549['Virus Added (uL) in 3mL total'], A549['RFP+ %'], color='tab:red')
ax.plot(df2['Virus Added (uL) in 2mL total'], df2['RFP+ %'], color='lightblue')



ax.legend(fontsize=14)

ax.tick_params(axis='both', which='major', labelsize=13)
ax.set_xlabel('Virus Added (uL)', fontsize=14)
ax.set_ylabel('RFP+ %', fontsize=14)

ax.spines[['right', 'top']].set_visible(False)
fig.tight_layout()
#fig.savefig('titer1.pdf')
```

# 9/19/24

CBE MUGIS results for A549-CBE6-NG-P2A-Puro cells (5 ug/mL puromycin; 2X transduction)

72 hours post-MUGIS transduction

```python
df = pd.read_csv('2024_9_19_CBE_MUGIS_A549.csv')[:-2]

df = df.rename(columns = {'Unnamed: 0':'Sample', 'singlets/singlets2/singlets3/RFP+ | Freq. of Parent (%)': 'RFP+ %', 'singlets/singlets2/singlets3/RFP+/GFP+ | Freq. of Parent (%)': 'GFP+ %'}).drop(columns = 'Unnamed: 3')
df

df['Cell Type'] = ['A549\nCBE6-NG']*3 + ['Uncolored\nControl'] + ['A549 WT']*3
df
```

```python
df_samps_only = df[df['Cell Type']!='Uncolored\nControl']

fig, ax = plt.subplots(1,2, figsize=(7,5), sharey=True, width_ratios=[3,2])

order = ['Uncolored\nControl', 'A549 WT', 'A549\nCBE6-NG']
sns.barplot(data = df, x='Cell Type', y='RFP+ %', ax=ax[0], color='tab:blue', linewidth=1, edgecolor='black', order=order)
sns.barplot(data = df_samps_only, x='Cell Type', y='GFP+ %', ax=ax[1], color='tab:blue', linewidth=1, edgecolor='black')
sns.stripplot(data = df, x='Cell Type', y='RFP+ %', ax=ax[0], color='tab:blue', linewidth=1, edgecolor='black', dodge=True, s=15, order=order)
sns.stripplot(data = df_samps_only, x='Cell Type', y='GFP+ %', ax=ax[1], color='tab:blue', linewidth=1, edgecolor='black', dodge=True, s=15)

ax[0].set_xlabel('')
ax[1].set_xlabel('')

ax[0].tick_params(axis='both', which='major', labelsize=15)
ax[1].tick_params(axis='both', which='major', labelsize=15)

ax[0].set_ylabel('RFP+ %', fontsize=14)
ax[1].set_ylabel('GFP+ %', fontsize=14)

ax[0].spines[['right', 'top']].set_visible(False)
ax[1].spines[['right', 'top']].set_visible(False)


fig.tight_layout()
#fig.savefig('MUGIS_CBE_A549_2024_9_19.pdf')
```
